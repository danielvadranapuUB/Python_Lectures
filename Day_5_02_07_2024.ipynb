```python
old_names = ["Sravan", "snehitha", "daniel", "Sukesh", "snehitha", ""]
new_names = []

for elements in old_names:

    if elements[0].islower():
        # elements[0] = 
        new_names.append(elements.capitalize())
    else:
        new_names.append(elements)
print(new_names)
```
